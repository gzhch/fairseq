```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
from scipy.stats import wasserstein_distance
```

```python
l = 1
h = 12
t = last_attn[l].squeeze().tolist()[h]
pd = pandas.DataFrame(t)
sns.heatmap(pd)
plt.show()
```

```python
for layer in range(24):
    fig, axs = plt.subplots(4, 16)
    fig.set_size_inches(32, 8)
    t_fft = fft_attn[layer].squeeze().tolist()
    t_rft = rft_attn[layer].squeeze().tolist()
    t_last = last_attn[layer].squeeze().tolist()
    t_first = first_attn[layer].squeeze().tolist()
    for i in range(16):
        pd1 = pandas.DataFrame(t_fft[i])
        pd2 = pandas.DataFrame(t_rft[i])
        pd3 = pandas.DataFrame(t_last[i])
        pd4 = pandas.DataFrame(t_first[i])
        g1 = sns.heatmap(pd1, ax=axs[0, i], xticklabels = False, yticklabels = False, cbar = False)
        g2 = sns.heatmap(pd2, ax=axs[1, i], xticklabels = False, yticklabels = False, cbar = False)
        g3 = sns.heatmap(pd3, ax=axs[2, i], xticklabels = False, yticklabels = False, cbar = False)
        g4 = sns.heatmap(pd4, ax=axs[3, i], xticklabels = False, yticklabels = False, cbar = False)
    plt.savefig('images/example_1/layer_'+str(layer)+'.png', dpi=100)
```

```python
def sim_basic(a, b):
    res = 0
    for i in range(len(a)):
        res += wasserstein_distance(a[i], b[i])
    return res
```

```python
for layer in range(24):
    a = fft_attn[layer].squeeze().tolist()
    b = rft_attn[layer].squeeze().tolist()
    c = last_attn[layer].squeeze().tolist()
    d = first_attn[layer].squeeze().tolist()
    res_ab = res_ac = res_bc = res_ad = res_bd = res_cd = 0
    for i in range(16):
        ab = sim_basic(a[i], b[i])
        ac = sim_basic(a[i], c[i])
        bc = sim_basic(b[i], c[i])
        ad = sim_basic(a[i], d[i])
        bd = sim_basic(b[i], d[i])
        cd = sim_basic(c[i], d[i])
        res_ab += ab
        res_ac += ac
        res_bc += bc
        res_ad += ad
        res_bd += bd
        res_cd += cd
    print(layer, res_ab, res_ac, res_bc, res_ad, res_bd, res_cd)
```
